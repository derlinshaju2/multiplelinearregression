```python
import pandas as pd
```

```python
d=pd.read_csv('https://raw.githubusercontent.com/codebasics/py/refs/heads/master/ML/2_linear_reg_multivariate/homeprices.csv')
```

```python
d
```

```python
d.isna().sum() #to see the null value
```

```python
a=d['bedrooms'].mean()
```

```python
a
```

```python
a.round()
```

```python
d=d.fillna({'bedrooms':a.round()})
```

```python
d
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(d['area'],d['price'])
```

```python
from sklearn.linear_model import LinearRegression
```

```python
li=LinearRegression()
```

```python
li.fit(d[['area','bedrooms','age']],d['price'])
```

```python
li.predict([[5000,5.0,10]])
```
